```python
"""
Programs analysis
"""
import pandas as pd

NUM_PROGRAM = 8
programs_df = []

for i in range(1, NUM_PROGRAM+1):
    program = pd.read_csv('../data/Program0%d.csv' % (i))
    print('Program %d' % (i))
    print('Episodes: %d' % (len(program)))
    print(program.columns)
    print(program.loc[:1]['Content'])
    programs_df.append(program)
```

```python
"""
Question analysis
"""

questions_df = pd.read_csv('../data/Question.csv')

print('Question')
print('Episodes: %d' % (len(questions_df)))
print(questions_df.columns)

print(questions_df.loc[:2]['Question'])

for i in range(6):
    print(questions_df.loc[:2]['Option%d' % (i)])
    print()
```

```python
"""
Set up jieba
"""

import jieba
jieba.set_dictionary('../data/dictionary.txt')
jieba.enable_parallel(16)
```

```python
"""
Define tokenizer, throw away space and tab
"""

invalid = ['', ' ', '\t']

def tokenize(sentences, is_question):
    token = []
    
    for sentence in sentences:
        tokenized_sentence = jieba.lcut(sentence, cut_all=True)
        no_space = [t for t in tokenized_sentence if t not in invalid]
        if no_space:
            if is_question:
                token.extend(no_space)
            else:
                token.append(no_space)
    return token
```

```python
"""
Tokenize the program
"""

from tqdm import tqdm_notebook

tokenized_programs = []
word2vec_train = []

for program in tqdm_notebook(programs_df):
    tokenized_program = []
    
    for index in range(len(program)):
        sentences = program.loc[index]['Content'].split('\n')
        tokens = tokenize(sentences, False)
        tokenized_program.append(tokens)
        word2vec_train.extend(tokens)
    
    tokenized_programs.append(tokenized_program)
```

```python
# The first sentence of the first episode of the first program
print(tokenized_programs[0][10]) 
```

```python
from tqdm import tqdm_notebook
tokenized_questions = []
n = len(questions_df)

for i in tqdm_notebook(range(n)):
    tokenized_question = []
    sentences = questions_df.loc[i]['Question'].split('\n')
    tokenized_question.append(tokenize(sentences, True))
    
    for j in range(6):
        option = questions_df.loc[i]['Option%d' % (j)]
        tokenized_option = [t for t in jieba.lcut(option, cut_all=True) if t not in invalid]
        tokenized_question.append(tokenized_option)
    
    tokenized_questions.append(tokenized_question)
```

```python
import numpy as np

np.save('../data/tokenized_programs.npy', tokenized_programs)
np.save('../data/tokenized_questions.npy', tokenized_questions)
np.save('../data/word2vec_train.npy', word2vec_train)
```

```python
"""
Analysis the question length
"""

count = 0
option_count = 0
tokenized_questions = np.load('../data/tokenized_questions.npy')

for question in tokenized_questions:
    count += len(question[0])
    for option in question[1:]:
        option_count += len(option)
    
print('average question length:', count/500)
print('average option length:', option_count/3000)
```

```python
"""
Extend the context, tried to get better training data
"""

from tqdm import tqdm_notebook
extended_programs = list()
for program in tqdm_notebook(tokenized_programs):
    extended_episode = list()
    for episode in tqdm_notebook(program):
        extended_sentence = list()
        zip_episodes = zip(episode[:-3], episode[1:-2], episode[2:-1])
        for s0, s1, s2 in zip_episodes:
            extended_sentence.append(s0 + s1 + s2)
        extended_episode.append(extended_sentence)
    extended_programs.append(extended_episode)
```

```python
np.save('../data/extended_programs.npy', extended_programs)
```
